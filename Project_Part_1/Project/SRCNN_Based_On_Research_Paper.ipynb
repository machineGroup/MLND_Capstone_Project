### Loading the images and generate sub_images for training

```python
images_folder = '../T91/'
# As per details given in the paper, I choose patch size as 33 on T91 images dataset. Chossing stride as 8 to get more sub-images for training
train_images=generate_training_dataset(images_folder, 91, 33, 8, 'T91_images')
```

We give 'scale_percent' as an input between 0 and 100. The image is downscaled by that factor and then using bicubic interpolation, it is brought back to the original size as a blurred image. The bicubic interpolation is done as a pre-processing step when we implement super-resolution for new images. This creates the set of images to be used as artificial training images against our actual training images generated from the T91 dataset.

### Generating the blurred images to be used for training the CNN using bicubic interpolation

```python
def get_interpolated_images(train_images, scale_percent):
    
    interpolated_images=np.zeros_like(train_images)
    
    for i in range(train_images.shape[0]):  
        
        img = train_images[i]
        
        scale_percent = scale_percent # percent of original size
        width = int(img.shape[1] * scale_percent / 100)
        height = int(img.shape[0] * scale_percent / 100)
        dim = (width, height)

        # Adding Gaussian Blur to the high resolution image
        img = cv2.GaussianBlur(img, (0, 0), 1, 1)
        
        low_resolution_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        
        width = int(img.shape[1])
        height = int(img.shape[0])
        dim = (width, height)
        
        # Bicubic interpolation preprocessing step
        interpolated_img = cv2.resize(low_resolution_img, dim, interpolation = cv2.INTER_CUBIC)

        interpolated_images[i]=interpolated_img  
    
    return interpolated_images
```

```python
# Getting the downscales images by a scaling factor of 3
interpolated_images=get_interpolated_images(train_images, 33.33)

# Check on size of images after processing
print(interpolated_images.shape)
print(train_images.shape)
```

Total 65895 images of blurred type are created which act as the training data while the high resolution images act as labels of data for training the convolutional neural network model Next we build the SRCNN model. 

### Building the CNN Model 

```python
from LearningModel import create_base_SRCNN_model
```

The SRCNN model has three convolutional layers: patch extraction layer, non-linear mapping layer and reconstruction layer.
The first layer extracts patches of size 9X9 from the images which are then mapped in a non-linear fashion to another set of filters in the second layer. The output filters from this layer are then used to generate the high-resolution image and compare it to the ground-truth image to calculate error. The PSNR (Peak Signal to Noise Ratio) is used as an evaluation metric for training the model. It is a metric based on Mean-Squared Error value between two images.

```python
# Specifying the input shape as 33X33X3
input_shape=train_images.shape[1:]

# The number of layers, number of filters in each layer and filter_size are as specified in the base case of the paper
model = create_base_SRCNN_model(input_shape, 9, 1, 5, 64, 32)
print(model.summary())
```

The above model is created using the values of number of filters and kernel_sizes mentioned for the base case in the paper.

### Setting the parameters for training

```python
# Inputs for training the CNN model
batch_size = 128
epochs = 50
```

```python
model.fit(interpolated_images, train_images, batch_size=batch_size, epochs=epochs, validation_split=0.2, shuffle=True)
```

## Model Evaluation

```python
# PSNR Metric Evaluation

def psnr(img1, img2, max_pixel):
    mse = np.mean( (img1 - img2) ** 2 )
    if mse == 0:
        return -1
    PIXEL_MAX = max_pixel
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
```

Let us visualize how well our predicted images are when compared to the actual label images and check the psnr scores for the same.

```python
# Get the PSNR Value averaged over an entire dataset

def get_average_psnr(model, interpolated_images, train_images):
    total_psnr=0

    for i in range(interpolated_images.shape[0]):
        
        if i%1000==0:
            print(i)
        
        predicted_image = model.predict(np.expand_dims(interpolated_images[i],axis=0))
        predicted_image = predicted_image[0]

        total_psnr+=psnr(train_images[i],predicted_image,1)
    
    average_psnr = total_psnr/interpolated_images.shape[0]
    
    return average_psnr
```

```python
print("Average PSNR for training dataset is {}".format(get_average_psnr(model, interpolated_images,train_images)))
```

```python
import keras.backend as K

def PSNR(y_true, y_pred):
    """
    PSNR is Peek Signal to Noise Ratio, which is similar to mean squared error.
    It can be calculated as
    PSNR = 20 * log10(MAXp) - 10 * log10(MSE)

    However, since we are scaling our input, MAXp = 1. Therefore 20 * log10(1) = 0.
    Thus we remove that component completely and only compute the remaining MSE component.
    """
    return -10. * K.log(K.mean(K.square(y_pred - y_true))) / K.log(10.)
```

```python
from keras.models import load_model
model = load_model('research_paper_based_model.h5', custom_objects={"PSNR": PSNR})
```
