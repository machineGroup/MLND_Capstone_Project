## SRCNN Model Training on Manga109 Dataset

Input Image Size: 32X32 (low-resolution)

Output Image Size: 64X64 (high-resolution)

An SRCNN Model is trained on the Manga109 Dataset by extracting sub-images of sizes 64X64, transforming them to get the corresponding low-resolution images and then training the CNN model for this set of data. This is the second part of the project and an extension of the proof of concept completed in Part1 based on a research paper. Kindly refer the GitHub repository https://github.com/soham239/MLND_Capstone_Project for the first half of the project. 

### Generating Training Data

```python
import math
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt                        
# %matplotlib inline
```

```python
from DataGenerator import generate_training_dataset, read_training_dataset
```

The Data Generator File contains the function which reads the patches of high resolution images, scales them to range 0 to 1 and stores them in the form of numpy arrays called 'train_images'. The images are of RGB format and the output size is 64X64 pixels.

```python
images_folder = '../Manga109/'

train_images=generate_training_dataset(images_folder, 109, 64, 48, 'Manga109_images')
```

```python
def get_interpolated_images(train_images, scale_percent):
    
    interpolated_images=np.zeros_like(train_images)
    
    for i in range(train_images.shape[0]):  
        
        if i%10000==0:
            print(str(i)+" images interpolated.")
        
        img = train_images[i]
        
        scale_percent = scale_percent # percent of original size
        width = int(img.shape[1] * scale_percent / 100)
        height = int(img.shape[0] * scale_percent / 100)
        dim = (width, height)

        # Adding Gaussian Blur to the high resolution image
        img = cv2.GaussianBlur(img, (0, 0), 1, 1)
        
        low_resolution_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        
        width = int(img.shape[1])
        height = int(img.shape[0])
        dim = (width, height)
        
        # Bicubic interpolation preprocessing step
        interpolated_img = cv2.resize(low_resolution_img, dim, interpolation = cv2.INTER_CUBIC)

        interpolated_images[i]=interpolated_img  
    
    return interpolated_images
```

```python
# Getting the downscales images by a scaling factor of 2
interpolated_images=get_interpolated_images(train_images, 50)

# Check on size of images after processing
print(interpolated_images.shape)
print(train_images.shape)
```

Total 41802 images of blurred type are created which act as the training data while the high resolution images act as labels of data for training the covolutional neural network model Next we build the SRCNN model. 

### Building the SRCNN model

```python
from LearningModel import create_base_SRCNN_model
```

The SRCNN model has three convolutional layers: patch extraction layer, non-linear mapping layer and reconstruction layer.
The first layer extracts patches of size 9X9 from the images which are then mapped in a non-linear fashion to another set of filters in the second layer. The output filters from this layer are then used to generate the high-resolution image and compare it to the ground-truth image to calculate error. The PSNR (Peak Signal to Noise Ratio) is used as an evaluation metric for training the model. It is a metric based on Mean-Squared Error value between two images.

```python
# Specifying the input shape as 33X33X3
input_shape=train_images.shape[1:]

# The number of layers, number of filters in each layer and filter_size are as specified in the base case of the paper
model = create_base_SRCNN_model(input_shape, 9, 3, 8, 128, 64)
print(model.summary())
```

### Training Process

```python
# Inputs for training the CNN model
batch_size = 128
epochs = 30
```

```python
model.fit(interpolated_images, train_images, batch_size=batch_size, epochs=epochs, validation_split=0.2, shuffle=True)
```

### Model Evaluation

```python
# PSNR Metric Evaluation

def psnr(img1, img2, max_pixel):
    mse = np.mean( (img1 - img2) ** 2 )
    if mse == 0:
        return -1
    PIXEL_MAX = max_pixel
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
```

Let us visualize how well our predicted images are when compared to the actual label images and check the psnr scores for the same.

```python
# Get the PSNR Value averaged over an entire dataset

def get_average_psnr(model, interpolated_images, train_images):
    total_psnr=0

    for i in range(interpolated_images.shape[0]):
        
        if i%10000==0:
            print(str(i) + " images done.")
        
        predicted_image = model.predict(np.expand_dims(interpolated_images[i],axis=0))
        predicted_image = predicted_image[0]

        total_psnr+=psnr(train_images[i],predicted_image,1)
    
    average_psnr = total_psnr/interpolated_images.shape[0]
    
    return average_psnr
```

```python
print("Average PSNR for training dataset is {}".format(get_average_psnr(model, interpolated_images,train_images)))
```

### Load the HDF5 model for Visualization of Results

```python
import keras.backend as K

def PSNR(y_true, y_pred):
    """
    PSNR is Peek Signal to Noise Ratio, which is similar to mean squared error.
    It can be calculated as
    PSNR = 20 * log10(MAXp) - 10 * log10(MSE)

    However, since we are scaling our input, MAXp = 1. Therefore 20 * log10(1) = 0.
    Thus we remove that component completely and only compute the remaining MSE component.
    """
    return -10. * K.log(K.mean(K.square(y_pred - y_true))) / K.log(10.)
```

```python
from keras.models import load_model
model = load_model('32_to_64_manga_updated.h5', custom_objects={"PSNR": PSNR})
```

### Results Visualization

Visualize the bicubic interpolated images and the SRCNN improvement over those images by placing them side-by-side.

```python
plt.figure(figsize=(15,40))

for i in range(10):
    
    image_num = i*100
    
    bicubic_image = interpolated_images[image_num]
    actual_image = train_images[image_num]
    predicted_image = model.predict(np.expand_dims(bicubic_image,axis=0))[0]
    
    plt.subplot(10,3,i*3+1)
    actual_image = cv2.normalize(actual_image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    plt.imshow((actual_image * 255).astype(np.uint8))
    plt.title(" Ground Truth Image ")
    
    plt.subplot(10,3,i*3+2)
    bicubic_image = cv2.normalize(bicubic_image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    plt.imshow((bicubic_image * 255).astype(np.uint8))
    plt.title(" Bicubic Image with PSNR : "+ str(round(psnr(bicubic_image, actual_image, 1),2)))
    
    plt.subplot(10,3,i*3+3)
    predicted_image = cv2.normalize(predicted_image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    plt.imshow((predicted_image * 255).astype(np.uint8))
    plt.title(" SRCNN Image with PSNR : "+ str(round(psnr(predicted_image, actual_image, 1),2)))
```

In most of the images, we can observe from the PSNR values that our SRCNN model has brought an improvement in the image over the bicubic interpolation step. The purpose of this exercise was to prove that the SRCNN model is a feasible and useful process to upscale the images and aid in the process of super-resolution.
